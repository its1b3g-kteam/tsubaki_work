# チャンネルの入れ替え

OpenCVを使って画像を読み込んだ場合、
色の順番は通常の「R,G,B」の順ではなく「B,G,R」の順番で読み込まれる

```python
#import
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#読み込み
img = cv2.imread("assets/imori.jpg")
```

```python
#画像の表示
plt.imshow(img)
#cv2.imshow("",img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#入れ替え
def change_channel(img):
    r = img[:,:,2].copy()
    b = img[:,:,0].copy()
    g = img[:,:,1].copy()
    
    img[:,:,0] = r
    img[:,:,1] = b
    img[:,:,2] = g
    
    return img
```

```python
out = change_channel(img)
plt.imshow(out)
#cv2.imshow("",out)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

# グレースケール化

R,B,Gにそれぞれ数値をかけて足し合わせる
gray = 0.2126 * r 0.7152 * g + 0.0722 * b

```python
def gray_scale(img):
    #floatに型を変換する
    img_ = img.astype(np.float)
    #RBGの取り出し
    r = img_[:,:,2]
    b = img_[:,:,0]
    g = img_[:,:,1]
    
    #gray scale
    gray = 0.2126 * r + 0.7152 * b + 0.0722 * b
    out = gray.astype(np.uint8)
    return out
```

```python
gray = gray_scale(img)

plt.imshow(gray)
```

# 二値化

画像を白黒にする
グレースケールでチャネルを1つにしたものを閾値を決めて0と255に変更する


```python
#グレースケールを二値化する
out_two = gray.copy()

#閾値 以下なら値を0に以上なら255に変更する
th = 128
out_two[out_two<th] = 0
out_two[out_two>= th] = 255
```

```python
plt.imshow(out_two)
#cv2.imshow("",out_two)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
