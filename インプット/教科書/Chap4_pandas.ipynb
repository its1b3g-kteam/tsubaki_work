# データ形成

```python
df = pd.read_excel("data/201704health.xlsx")
```

```python
df.head()
```

条件抽出

```python
#TrueかFalseかをSeriesで返却する
df["歩数"] >= 10000
```

```python
#歩数が10000以上のデータのみ抽出
df_selected = df[df["歩数"] >= 10000]
df_selected
```

```python
df_selected.shape
```

```python
#queryメソッドを使用する場合
#df.query("条件")
df.query('歩数 >= 10000 and 摂取カロリー <= 1800')
```

# データ型の変換

```python
#dtypes dataframeの各カラムのデータ型を一括で表示する
df.dtypes
```

```python
#applyメソッド　新しくカラムの追加
df.loc[:,'date'] = df.loc[:,'日付'].apply(pd.to_datetime)
df.head()
```

```python
#astypeメソッド　型の変換
df.loc[:,"摂取カロリー"] = df.loc[:,"摂取カロリー"].astype(np.float32)
```

indexのセット

```python
#indexにセット
df = df.set_index("date")
```

```python
df.head()
```

ソート

```python
#並べ替え sort_values(by = カラム)
#defaultは昇順
df.sort_values(by = "歩数").head(10)
```

```python
#降順の場合 ascending=False
df.sort_values(by = "歩数",ascending=False).head(10)
```

カラムの削除

```python
#カラムの場合、axis=1が必要
df = df.drop("日付",axis=1)
```

```python
df.tail()
```

組み合わせデータの挿入

```python
df.loc[:,"歩数/カロリー"] = df.loc[:,"歩数"] / df.loc[:,"摂取カロリー"]
```

```python
df.head()
```

```python
#関数の作成
def exercise_judge(ex):
    if ex <= 3.0:
        return "Low"
    elif 3.0 < ex < 6.0:
        return "Mid"
    else:
        return "High"
```

```python
df.loc[:,"運動指数"] = df.loc[:,"歩数/カロリー"].apply(exercise_judge)
df.head()
```

ダミー化

```python
#get_dumies カテゴリ変数をダミー化
#prefixはカラム名の先頭に付与する名前
df_moved = pd.get_dummies(df.loc[:,"運動指数"],prefix="運動")
df_moved.head()
```
