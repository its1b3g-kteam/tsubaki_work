# Matplotlib

```python
import matplotlib.pyplot as plt
```

# MATLABスタイル

```python
#データ用意
x = [1,2,3]
y = [2,4,9]

plt.plot(x,y)
plt.title("MATLAB-style")

plt.show()
```

# オブジェクト指向スタイル

    オブジェクト指向スタイルは描画オブジェクトに対してサブプロットを追加、
    サブプロットに対してグラフを描画する

```python
x = [1,2,3]
y = [2,4,9]

#描画オブジェクト(fig)サブプロット(ax)を生成
fig,ax = plt.subplots()#1つの描画オブジェクトを生成し、その中に1つのサブプロットを配置

ax.plot(x,y)
ax.set_title("OOP-style")

plt.show()
```

描画オブジェクト

    matplotlibでグラフを描画するには、描画オブジェクト(figure)を生成し、
    その中の1つ以上のサブプロット(subplot)を配置

```python
fig,axes = plt.subplots(2) #2つのサブプロットを配置
plt.show()
```

```python
fig,axes = plt.subplots(2,2) #2行2列のサブプロットを配置
plt.show()
```

各引数はキーワード引数nrows,ncolsでも指定が可能

```python
fig ,axes = plt.subplots(ncols=2)
plt.show()
```

グラフのスタイル

    matplotlib.styleを使用するとグラフの表示スタイルが指定できる
    有効なスタイル名の一覧はmatploylib.style.availableで取得できる(23種類)
    https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html

```python
#グラフのスタイルにclassicを指定
matplotlib.style.use("classic")

fig,ax = plt.subplots()
ax.plot([1,2])

plt.show()
```

```python
#タイトル
fig,axes = plt.subplots(ncols=2)

#サブプロットにタイトル指定
axes[0].set_title("subplot title 0")
axes[1].set_title("subplot title 1")

#描画オブジェクトにタイトルを設定
fig.suptitle("figure title")

plt.show()
```

```python
fig,ax = plt.subplots()

ax.set_xlabel("x label")#x軸のラベル
ax.set_ylabel("y label")#y軸のラベル

plt.show()
```

```python
fig,ax = plt.subplots()

#凡例用のラベルを設定
ax.plot([1,2,3],[2,4,9],label="legend label")
ax.legend(loc = "best")#凡例を表示
#凡例の位置はlocで指定が可能
#lower right　右下
#upper left　左上　等

plt.show()
```

# グラフ

折れ線グラフ

```python
#plot 折れ線グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]

ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

棒グラフ

```python
#bar 棒グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]
title = ["spam","ham","egg"]
ax.bar(x,y,tick_label = title)#tick_labelで各棒グラフにタイトルをつけることができる
#ax.barh(x,y,tick_label = title) 横向きに表示したいときはbarhを使用する
plt.show()
```

複数並べる場合

```python
#
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ["spam","ham","egg"]

width = 0.4#棒グラフの幅

ax.bar(x,y1,width=width,tick_label = labels,label="y1")

#幅をずらして棒グラフを描画
x2 = [num + width for num in x]
ax.bar(x2,y2,width=width,label="y2")

ax.legend()
plt.show()
```

```python
# 積み上げの場合
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ["spam","ham","egg"]

#合計値
y_total = [num1 + num2 for num1,num2 in zip(y1,y2)]

#y1とy2の合計した棒グラフを表示
ax.bar(x,y_total,tick_label = labels,label="y1")
ax.bar(x,y2,label="y2")

ax.legend()
plt.show()
```

散布図

```python
import numpy as np

fig ,ax = plt.subplots()

#ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x,y)
plt.show()
```

```python
fig,ax = plt.subplots()

#ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10],y[0:10],marker = 'v' , label = 'triangle down') #下向き

ax.scatter(x[10:20],y[10:20],marker = '^', label = 'triangle up') #上向き

ax.scatter(x[20:30],y[20:30],marker = 's', label = 'square') #正方形

ax.scatter(x[30:40],y[30:40],marker = '*', label = 'star') #星型

ax.scatter(x[40:50],y[40:50],marker = 'x', label = 'x') #X

plt.legend()

plt.show()
```

ヒストグラム

```python
#データ生成
np.random.seed(123)
mu = 100 #平均値
sigma = 15 #標準偏差
x = np.random.normal(mu,sigma,1000)

fig,ax = plt.subplots()

#ヒストグラムを描画
n,bins,patches = ax.hist(x)
plt.show()
```

```python
#度数
print("nは度数(要素数)",n)

print("binsはビンの境界値",bins)

print("patchesはビンを描画する情報",patches)
```

```python
fig,ax = plt.subplots()

ax.hist(x,bins=25) #ビンの数を指定して描画

plt.show()
```

```python
fig,ax = plt.subplots()

#横向きのヒストグラムを描画
ax.hist(x,orientation = 'horizontal')

plt.show()
```

```python
#データ生成
np.random.seed(123)
mu = 100 #平均値
x0 = np.random.normal(mu,20,1000)

#異なる標準偏差でデータを生成
x1 = np.random.normal(mu,15,1000)
x2 = np.random.normal(mu,10,1000)

fig,ax = plt.subplots()

labels = ['x0','x1','x2']
#3つのデータのヒストグラムを描画
ax.hist((x0,x1,x2), label = labels)
ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()

labels = ['x0','x1','x2']
#積み上げたヒストグラムを描画 stacked = True
ax.hist((x0,x1,x2),label = labels,stacked = True)
ax.legend()
plt.show()
```

箱ひげ図

```python
#異なる標準偏差でデータを生成
np.random.seed(123)
x0 = np.random.normal(0,10,500)
x1 = np.random.normal(0,15,500)
x2 = np.random.normal(0,20,500)

fig,ax = plt.subplots()
labels = ['x0','x1','x2']
ax.boxplot((x0,x1,x2),labels = labels) #箱ひげ図を描画

plt.show()
```

```python
#横向きの場合 vert = False
fig,ax = plt.subplots()
labels = ['x0','x1','x2']
ax.boxplot((x0,x1,x2),labels = labels,vert = False) #箱ひげ図を描画

plt.show()
```

    円グラフ(初見)
    pieメソッド

```python
labels = ["spam","ham","egg"]
x = [10,3,1]

fig,ax = plt.subplots()

ax.pie(x,labels = labels)
ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()

ax.pie(x,labels = labels)
ax.axis('equal') #アスペクト比を保持
ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()
#pieはデータの少ない順に通常は描画する

#startangleは開始時点を決める　デフォルトは3時方向　90は12時から
#counterclock=Falseにすると時計周りに表示
ax.pie(x,labels = labels,startangle=90,counterclock = False)#上から時計まわり
ax.axis('equal') #アスペクト比を保持
ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()

#autopct は桁数
ax.pie(x,labels = labels,startangle=90,counterclock=False,shadow=True,autopct="%1.2f%%")
ax.axis("equal")

plt.show()
```

```python
#切り出す幅を指定
explode = [0,0.2,0] #1番目の要素(ham)を切り出す

fig,ax = plt.subplots()
#explodeは要素は切り出す
ax.pie(x,labels = labels,startangle=90,counterclock=False,shadow=True,autopct="%1.2f%%",explode =explode)

ax.axis("equal")

plt.show()
```

複数のグラフを組み合わせる

# スタイル

    色の指定は
    colorで塗りつぶし
    edgecolorで枠線を色づける

```python
fig,ax = plt.subplots()

#線の色を名前で指定
ax.plot([1,3],[3,1],label = "aqua",color = "aqua")
#16進数のRBGで指定
ax.plot([1,3],[1,3],label = "#0000FF",color = "#0000FF")
#RGBAをfloatで指定
ax.plot([1,3],[2,2],label = "(0.1,0.2,0.5,0.3)",color = (0.1,0.2,0.5,0.3))

ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()

ax.bar([1],[3],color = "aqua")
ax.bar([2],[4],color = "aqua",edgecolor = "red")

plt.show()
```

```python
fig ,ax = plt.subplots()
#線の幅 linewidth

#5.5ポイントの幅の線で描画
ax.plot([1,3],[3,1],linewidth = 5.5,label = "5.5")
#10ポイントの幅の線で描画
ax.plot([1,3],[1,3],linewidth = 10,label ="10")
ax.legend()

plt.show()
```

```python
fig ,ax = plt.subplots()

#破線で描画
ax.plot([1,3],[3,1],linestyle = "--",label = "dashed")

ax.plot([1,3],[1,3],linestyle = "-.",label = "dashdot")

ax.plot([1,3],[2,2],linestyle = ":",label = "dotted")

ax.legend()

plt.show()
```

フォント

```python
fig ,ax = plt.subplots()

ax.set_xlabel("xlabel",family = "fantasy",size=20,weight="bold")
ax.set_ylabel("ylabel",family = "cursive",size=40,weight = "light")
plt.show()
```

```python
#辞書を使用してまとめて指定もできる
fontdict = {
    "family":"fantasy",
    "size":20,
    "weight":"normal",
}

fig,ax = plt.subplots()

ax.set_xlabel("xlabel",fontdict = fontdict)
ax.set_ylabel("ylabel",fontdict = fontdict)

#x座標,y座標,text,文字サイズ
ax.text(0.2,0.4,"Text",size=20)
```
