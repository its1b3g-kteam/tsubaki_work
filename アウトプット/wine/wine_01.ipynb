```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#読み込み
wine_data = pd.read_csv("train.tsv",delimiter='\t')
wine_test_data = pd.read_csv("test.tsv",delimiter='\t')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#モデル作成　評価
X = wine_data.drop(["id","Y"],axis=1)
Y = wine_data["Y"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
clf = RandomForestClassifier(n_estimators=1000,max_depth=3,random_state=0)

clf.fit(X_train,y_train)
```

```python
#predict
result = clf.predict(X_test)
```

```python
#score
score = clf.score(X_test,y_test)

print(score)
```

```python
y_test_data = wine_test_data.drop(["id"],axis=1)
```

```python
result = clf.predict(y_test_data)
```

```python
# 提出用に加工
np_id = wine_test_data['id'].values
dd=pd.DataFrame({"id":np_id, "ans":result})
dd.to_csv("result01_nomal.csv",header=False,index=False)
dd
```
